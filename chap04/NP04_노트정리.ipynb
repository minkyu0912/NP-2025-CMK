__파이썬 자료 구조__
=============

* * *
## __문자열(String)__
문자열은 문자나 단어등으로 구성된 문자의 집합을 표현하는 자료구조이다.

1. __문자열 정의__
     <br> • 다수의 문자를 표현하기 위한 자료구조
     <br> • '' 또는 ""로 둘러싼다: "abc", 'hello'
     <br> • '''(""") 부터 '''(""")까지 문자: 문자열이 두 줄 이상일 때 주로 사용
     
2. __문자열의 인덱싱과 슬라이싱__
     <br> • 문자를 개별적으로 사용
     <br> • 문자열을 잘라낸다

3. __문자열 연산__

4. __문자열 관련 함수__

## __리스트(list)__
리스트는 여러가지 자료를 하나로 모아서 표현할 수 있는 자료구조이다.

1. __리스트를 만드는 방법__

2. __리스트의 인덱싱과 슬라이싱__

3. __연산, 수정, 삭제 방법__

4. __리스트관련함수__


* * *
## __튜플(tuple)__
튜플은 리스트와 같이 여러가지 자료를 모아서 하나로 표현할 수 있으나 요소를 수정, 삭제할 수 없는 자료구조이다

1. __튜플 만들기__

2. __튜플의 더하기와 곱하기__

3. __튜플의 인덱싱과 슬라이싱__


* * *
## __딕셔너리(dictionary)__
딕셔너리는 사전과 같이 대응 관계를 나타내는 자료구조이다.
 
1. __딕셔너리 만들기__

2. __딕셔너리 요소 추가, 삭제하기__

3. __딕셔너리 사용하기__

4. __딕셔너리 관련 함수__
